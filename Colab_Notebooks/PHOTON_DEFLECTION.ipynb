```python
from sympy import symbols, integrate, Rational, oo, simplify

r, r_0, b, R_s, G, M, c, t, v = symbols('r r_0 b R_s G M c t v', positive=True, real=True)
kappa_0 = symbols('kappa_0', positive=True)

print("=" * 70)
print("COMPARISON: MASSIVE PARTICLE vs PHOTON DEFLECTION")
print("=" * 70)

# Massive particle: Phi_mass = (1/2) R_s c^2 / r
print("\n--- Massive particle (dual-axis, Phi = R_s c^2 / (2r)) ---")
integrand_mass = (R_s * c**2 / 2) * b / (b**2 + v**2 * t**2)**Rational(3,2)
dv_mass = integrate(integrand_mass, (t, -oo, oo))
alpha_mass = simplify(dv_mass / v)
print(f"  alpha_mass = {alpha_mass}")
print(f"  = R_s*c^2 / (b*v^2)")

# For v = c
alpha_mass_vc = alpha_mass.subs(v, c)
print(f"\n  At v = c: alpha_mass = {simplify(alpha_mass_vc)}")

# Photon: Phi_gamma = R_s c^2 / r (doubled, single-axis)
print("\n--- Photon (single-axis, Phi = R_s c^2 / r) ---")
integrand_photon = R_s * c**2 * b / (b**2 + c**2 * t**2)**Rational(3,2)
dv_photon = integrate(integrand_photon, (t, -oo, oo))
alpha_photon = simplify(dv_photon / c)
print(f"  alpha_photon = {alpha_photon}")

# Ratio
ratio = simplify(alpha_photon / alpha_mass_vc)
print(f"\n  alpha_photon / alpha_mass(v=c) = {ratio}")
print("  => Light deflection is TWICE the massive-particle-at-c deflection")

# Express in WILL RG variables
print("\n\n" + "=" * 70)
print("FULL WILL RG EXPRESSION WITH DEPENDENCY CHAIN")
print("=" * 70)
print()
print("LIGHT DEFLECTION ANGLE:")
print()
print("  alpha = 2 * kappa_0^2")
print()
print("where kappa_0^2 = R_s / r_0 (field equation at closest approach)")
print()
print("DEPENDENCY CHAIN:")
print()
print("Level 0 (fundamental):")
print("  kappa_0 = sin(theta_2)  on S^2 carrier")
print("  kappa_X0 = cos(theta_2) = sqrt(1 - kappa_0^2)")
print()
print("Level 1 (field equation):")
print("  kappa_0^2 = R_s / r_0")
print()
print("Level 2 (system scale):")
print("  R_s = kappa^2 * a                    (from orbital data)")
print("  R_s = 2*G*M / c^2                    (from mass data)")
print("  R_s = r1*r2/(r2-r1)*(beta1^2-beta2^2) (two-point method)")
print()
print("Level 3 (output):")
print("  alpha = 2 * R_s / r_0")
print("        = 4 * G * M / (c^2 * r_0)")

# Numerical verification: Sun grazing
print("\n\n" + "=" * 70)
print("NUMERICAL VERIFICATION: SUN-GRAZING PHOTON")
print("=" * 70)

G_val = 6.674e-11  # m^3 kg^-1 s^-2
M_sun = 1.989e30   # kg
c_val = 2.998e8     # m/s
R_sun = 6.957e8     # m

R_s_sun = 2 * G_val * M_sun / c_val**2
print(f"\n  R_s(Sun) = 2GM/c^2 = {R_s_sun:.4f} m = {R_s_sun/1000:.4f} km")

kappa_0_sq = R_s_sun / R_sun
print(f"  kappa_0^2(Sun surface) = R_s/R_sun = {kappa_0_sq:.6e}")
print(f"  kappa_0(Sun surface) = {kappa_0_sq**0.5:.6e}")

alpha_rad = 2 * R_s_sun / R_sun
alpha_arcsec = alpha_rad * 206265  # convert to arcseconds
print(f"\n  alpha = 2 * kappa_0^2 = {alpha_rad:.6e} rad")
print(f"        = {alpha_arcsec:.4f} arcseconds")
print(f"\n  GR prediction: 1.7512 arcseconds")
print(f"  Eddington 1919 measurement: 1.75 ± 0.06 arcseconds")
print(f"  VLBI modern: 1.7520 ± 0.0002 arcseconds")
```
