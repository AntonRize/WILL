```python
# 1. Install Whisper and the Interface tool (Gradio)
!pip install git+https://github.com/openai/whisper.git gradio
!apt install ffmpeg

import gradio as gr
import whisper

# 2. Load the "Medium" model (Good balance of speed/accuracy)
print("Loading AI model... this takes about 30 seconds...")
model = whisper.load_model("medium")

# 3. Define the transcription function
def transcribe_audio(audio_file):
    if audio_file is None:
        return "Please upload an audio file."
    print(f"Transcribing: {audio_file}")
    # Transcribe
    result = model.transcribe(audio_file)
    return result["text"]

# 4. Launch the interface
print("Starting the App... Click the 'gradio.live' link below when it appears!")
iface = gr.Interface(
    fn=transcribe_audio,
    inputs=gr.Audio(type="filepath", label="Upload M4A File"),
    outputs=gr.Textbox(label="Transcription"),
    title="Free Whisper Transcriber",
    description="Upload your M4A file and wait for the text."
)

iface.launch(share=True, debug=True)
```
