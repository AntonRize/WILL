```python
import math

# --- PHYSICAL CONSTANTS & SYSTEM INPUTS ---
c = 299792458.0      # Speed of light (m/s)
T_0 = 27906.98       # Initial orbital period (s)
delta_T = -0.0000765 # Annual period decay (s/year)
e = 0.6171334        # Eccentricity
phi_dot_deg_yr = 4.22659 # Precession per year

# --- DERIVE INITIAL GEOMETRIC STATE (State 0) ---
# Calculate precession per orbit to find beta_0
orbits_per_year = (365.25 * 86400) / T_0
delta_phi_rad = math.radians(phi_dot_deg_yr / orbits_per_year)
beta_0 = math.sqrt((delta_phi_rad * (1 - e**2)) / (6 * math.pi))

# Anchor the system scale (R_s) - This remains invariant for the system
R_s = (T_0 * c / math.pi) * (beta_0**3)

# Initial derived geometry
a_0 = R_s / (2 * beta_0**2)
v_0 = beta_0 * c

# --- DERIVE DECAYED GEOMETRIC STATE (State 1) ---
T_1 = T_0 + delta_T

# Calculate new beta_1 based on the updated period and invariant R_s
# Relation: beta = ( (pi * Rs) / (T * c) )^(1/3)
beta_1 = ((math.pi * R_s) / (T_1 * c))**(1/3)

# Calculate new geometric state
a_1 = R_s / (2 * beta_1**2)
v_1 = beta_1 * c

# --- CALCULATE GEOMETRIC SHIFTS ---
delta_a = a_1 - a_0
delta_v = v_1 - v_0
delta_beta = beta_1 - beta_0

print("=== WILL R.O.M. ORBITAL DECAY ANALYSIS ===")
print(f"System Anchor (R_s): {R_s:.4f} m")
print(f"Initial Period (T_0): {T_0:.6f} s")
print(f"Decay (Delta T): {delta_T:.8f} s/year")
print("-" * 40)
print(f"Initial State (a_0): {a_0:.4f} m")
print(f"Decayed State (a_1): {a_1:.4f} m")
print(f"Annual Orbital Shrinkage (Delta a): {delta_a:.6f} meters/year")
print("-" * 40)
print(f"Initial Velocity (v_0): {v_0:.6f} m/s")
print(f"Decayed Velocity (v_1): {v_1:.6f} m/s")
print(f"Annual Velocity Increase (Delta v): {delta_v:.6f} m/s")
print("-" * 40)
print(f"Relational Invariant Check (beta^2 * a): {beta_1**2 * a_1:.4f} (Should match R_s/2: {R_s/2:.4f})")
```
