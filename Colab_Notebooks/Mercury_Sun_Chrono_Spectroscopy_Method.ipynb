```python
import math

# INPUTS (Inter-Galactical Invariants & Local Standards)
theta_sun = 0.0046524
T_M = 87.969
T_E = 365.25
e_obs = 0.2056
z_sun = 2.1224e-6
c = 299792458
T_M_sec = T_M * 86400

# PIPELINE ALGEBRA
# 1. Relational Scale Factor
T_ratio = T_M / T_E
R_ratio = theta_sun / ((T_ratio**(2/3)) * (1 - e_obs))

# 2. Local Gravitational Projection (at perihelion)
kappa_p_sq = (1 - (1 / (1 + z_sun)**2)) * R_ratio

# 3. Local Kinematic Projection (at perihelion)
beta_p_sq = (kappa_p_sq / 2) * (1 + e_obs)

# 4. Anomalous Precession
delta_phi_rad = (3 * math.pi * kappa_p_sq) / (1 + e_obs)
orbits_per_century = (365.25 / T_M) * 100
delta_phi_arcsec_cy = delta_phi_rad * (180 * 3600 / math.pi) * orbits_per_century

# 5. Absolute System Scale
# Reconstructing global invariants from perihelion states
W = 0.5 * (kappa_p_sq - beta_p_sq)
kappa_global_sq = 4 * W
beta_global = math.sqrt(2 * W)
R_s = T_M_sec * c * (kappa_global_sq * beta_global) / (2 * math.pi)

# OUTPUT
print("--- CALCULATED VALUES ---")
print(f"R_ratio = {R_ratio:.9f}")
print(f"kappa_p^2 = {kappa_p_sq:.6e}")
print(f"beta_p^2 = {beta_p_sq:.6e}")
print(f"Delta_phi (rad/orbit) = {delta_phi_rad:.6e}")
print(f"Delta_phi (arcsec/century) = {delta_phi_arcsec_cy:.4f}")
print(f"R_s (meters) = {R_s:.1f}")
```
