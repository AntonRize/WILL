```python
import math

# --- OBSERVED DATA (From Earth's perspective) ---
# 1. Chronometric anchor
T_years = 50.1284
T = T_years * 365.25 * 86400  # Convert to seconds

# 2. Kinematics (Spectroscopy)
K_total = 10077.0  # Total spectroscopic semi-amplitude (m/s)
c = 299792458.0    # Speed of light (m/s)

# 3. Visual orbital geometry (Astrometry)
e = 0.5914
i_rad = math.radians(136.336)
theta_orb_arcsec = 7.4957
theta_orb = theta_orb_arcsec / 206265.0  # Convert to radians

# 4. Visual geometry of Sirius A (Interferometry)
theta_A_mas = 3.02
theta_A = (theta_A_mas / 1000.0) / 206265.0  # Convert to radians

print("=== WILL R.O.M. SIRIUS SYSTEM DECODING ===")

# STEP 1: Extract global kinematic projection (beta)
# K = beta * c * (sin(i) / sqrt(1 - e^2)) -> invert to find beta
beta = (K_total / c) * (math.sqrt(1 - e**2) / abs(math.sin(i_rad)))
print(f"1. Global kinematic projection (beta): {beta:.4e}")

# STEP 2: Absolute system scale (R_s)
# Derived strictly from time and kinematics, bypassing G and M
R_s = (T * c / math.pi) * (beta**3)
print(f"2. Absolute system scale (R_s): {R_s:.2f} meters")

# STEP 3: Absolute semi-major axis of the orbit (a)
a = R_s / (2 * beta**2)
print(f"3. Semi-major axis (a): {a:.4e} meters")

# STEP 4: Absolute distance to the system (D)
# Close the absolute scale onto the apparent orbital angle
D = a / theta_orb
print(f"4. Distance to Sirius (D): {D:.4e} meters ({(D / (3600*24*365.25*c)):.2f} light years)")

# STEP 5: Physical radius of Sirius A (R_A)
# Project the calculated distance onto the apparent size of the star
R_A = D * theta_A
print(f"5. Physical radius of Sirius A: {R_A:.4e} meters ({(R_A / 695700000):.2f} solar radii)")
```
